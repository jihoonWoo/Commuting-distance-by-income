## Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Read data

```python
input_path = '../../DATA/Commuting-Distance-by-Personal-Characteristics/generated_data/'
input_file = 'transformed_data.xlsx'
```

```python
df = pd.read_excel(input_path + input_file, usecols=[1, 2, 3])
df
```

The generated data contains the information about commuting distance in Australia in 2016. Column 'Age group(AGEP)' indicate different age groups, while 'Average commuting distance (kilometres)' column shows the average commuting distance for these age groups. The average commuting distance in Australia is 15.98km where age group 30's has the highest average commuting distance (16.59).   

## Visualize data

Here, we'd like to visualize the generated data. Below code demonstrates how to visualize the data by using Matplot library.

**Matplot** : Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')

x = list(df['Age group (AGEP)'])[:-1]
AVG_distances = list(df['Average commuting distance \n(kilometres)'])[:-1]

x_pos = [i for i, _ in enumerate(x)]
plt.figure(figsize=(22,14))
plt.bar(x_pos, AVG_distances, color='#F86051')
plt.xlabel("Age groups", fontsize=22)
plt.ylabel("Kilometres", fontsize=22)
plt.title("Average commuting distance (km) by age", fontsize=28)

for index, value in enumerate(AVG_distances):
    plt.text(index - 0.22, value - 0.8, str(round(value, 2)), fontsize=18)

plt.xticks(x_pos, x, fontsize=18, rotation = 90)

plt.savefig('img/Commuting distance by age.pdf')

plt.show()
```

The above figure shows people aged between 20 - 69 years report a similar average daily commuting distance of ( between 16.56 and 16.59km). While younger people aged  between 15-19 years  have the lowest average commuting distance (11.3km), followed by people aged between 70 -79 years (13.18km) 

## Insight

**People aged between 30-39 years have the highest average commuting distance (16.59km) among the age groups in Australia in 2016.**

```python
value = float(df['Average commuting distance \n(kilometres)'].loc[df['Age group (AGEP)'] == '30-39 years'])
print('People aged between 30-39 years have the highest average commuting distance ({}km) among the age groups in Australia in 2016.'.format(round(value, 2 )))
```
