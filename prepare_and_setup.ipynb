## Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

## Read data

```python
input_path = '../../DATA/Commuting-Distance-by-Personal-Characteristics/generated_data/'
input_file = 'transformed_data.xlsx'
```

```python
df = pd.read_excel(input_path + input_file, usecols=[1, 2, 3], sheet_name='income')
df
```

The generated data contains the information about commuting distance in Australia in 2016. Column 'Weekly Income (INCP)' includes different weekly income groups, while 'Average commuting distance (kilometres)' column shows the average commuting distance for these income groups. The average commuting distance in Australia is 15.98km where weekly income group 2,000 - 2,999 has the highest average commuting distance (20.01).   

## Visualize data

Here, we'd like to visualize the generated data. Below code demonstrates how to visualize the data by using Matplot library.

**Matplot** : Matplotlib is a comprehensive library for creating static, animated, and interactive visualizations in Python.

```python
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
plt.style.use('ggplot')

x = list(df['Weekly Income (INCP)'])
AVG_distances = list(df['Average commuting distance \\n(kilometres)'])

x_pos = [i for i, _ in enumerate(x)]
plt.figure(figsize=(22,14))
plt.bar(x_pos, AVG_distances, color='#56C060')
plt.xlabel("Income", fontsize=22)
plt.ylabel("Kilometres", fontsize=22)
plt.title("Average commuting distance (km) by Income", fontsize=28)
plt.ylim((0, 25))   # set the ylim to bottom, top
for index, value in enumerate(AVG_distances):
    plt.text(index-0.4, value - 1, str(round(value, 2)), fontsize=18)

plt.xticks(x_pos, x, fontsize=18, rotation=90)

plt.savefig('img/Commuting distance by income.pdf')

plt.show()
```

The increased average daily commuting distance shows a positive correlation with increased average weekly income. Interestingly people earning in excess of \$3000 per week report a lower average daily commute than people in the previous four income brackets.
